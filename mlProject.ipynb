```python
import pandas as pd
```

```python
file_path = '/Users/eriksundstrom/FirstMLProject/melb_data.csv'
melbourne_data = pd.read_csv(file_path) #reading data and storing it in a dataframe called melbourne_data
melbourne_data.describe() #prints a summary of the data the data
#count shows how many rows have non-missing values
```

```python
melbourne_data.head() #prints the first 5 rows
```

```python
print(melbourne_data.shape) #(rows, coloumns)
```

```python
melbourne_data.dropna(axis=0) #drops entries with missing values
```

```python
y = melbourne_data.Price #what we want to predict
```

```python
mel_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude'] #legg til building area!
X = melbourne_data[mel_features]
X.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
print("the acutal values were: ")
print(y.head())
```
